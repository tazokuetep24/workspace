Welche der folgenden Aussagen über funktionale Abhängigkeiten stimmen?

- __c:__ Es gibt feste Regeln, welcher Schlüsselkandidat als Primärschlüssel ausgewählt wird.<br>
- __d:__ Wenn $A \to B$ und $B \to C$ gelten, dann gilt auch $A \to C$.<br>
- __e:__ Die Menge aller Attribute ist immer ein Schlüsselkandidat.<br>
- __f:__ Die Menge aller Attribute ist immer ein Superschlüssel.<br>
- __g:__ Wenn ein Attribut auf der rechten Seite von keiner Regel vorkommt, muss es in jedem Schlüsselkandidaten enthalten sein.<br>
- __h:__ Die Attributhülle einer nicht leeren Attributmenge kann leer sein.<br>
